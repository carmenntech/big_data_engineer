```python
import pandas as pd
ruta = 'paro2021.csv'
df = pd.read_csv(ruta, sep=';')


# Asegurarse de que la columna de fecha está en formato datetime (reemplaza 'fecha' con el nombre de tu columna de fecha)
df['FECHA'] = pd.to_datetime(df['FECHA'])

# Extraer el mes y añadirlo como una nueva columna
df['mes'] = df['FECHA'].dt.month

df.head()
```

```python
import matplotlib.pyplot as plt

x = df['mes']
y = df['TOTAL PARADOS']

plt.figure()
plt.plot(x, y, marker='o', linestyle='-', color='b')
plt.show()

```

```python


# Graficar
fig, ax = plt.subplots()

ax.bar(x, y, width=0.8, edgecolor="white", linewidth=0.7)

ax.set()

ax.set_xlabel('Mes')
ax.set_ylabel('Valor')
ax.set_title('Valores por Mes')

plt.show()
```
