```python
import pandas as pd

ruta = 'paro2005_2023.csv'
df = pd.read_csv(ruta, sep=';')

df.head()
```

```python
# Asegurarse de que la columna de fecha está en formato datetime (reemplaza 'fecha' con el nombre de tu columna de fecha)
df['FECHA'] = pd.to_datetime(df['FECHA'])

# Extraer el mes y añadirlo como una nueva columna
df['mes'] = df['FECHA'].dt.month
df['anio'] = df['FECHA'].dt.year

df.head()
```

```python
import matplotlib.pyplot as plt

x = df['anio']
y = df['TOTAL PARADOS']

plt.figure()
plt.plot(x, y, marker='o', linestyle='-', color='b')
plt.show()
```
